```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as image
```

```python
img = image.imread("rose.jpg") # Only jpg images
w, h, d = img.shape
pixels = np.reshape(img, (w*h, d))
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=5, random_state=42).fit(pixels) # Change n_cluster size to get more or less colors
palette = np.uint8(model.cluster_centers_)
```

```python
_, ax = plt.subplots(1,2, figsize=(20,20))
ax[0].imshow(img)
ax[0].set_title("Original Image", fontsize=20)
ax[1].imshow([palette])
ax[1].set_title("Color Palette", fontsize=20)
plt.show()
```
