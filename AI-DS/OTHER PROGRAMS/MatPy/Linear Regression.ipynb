```python
sns.pairplot(advert, x_vars = ['TV', 'Radio', 'Newspaper'], y_vars = 'Sales',
             height=10, aspect=0.7, kind='reg')
```

```python
# Doing Linear Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = advert[['Newspaper']] # can be replaced with TV, Radio or Newspaper 
y = advert.Sales

# splitting dataset into Train & Test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_train.shape)

linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Interpretting Model Coefficients
print("Model coefficient value: ", linreg.coef_) # model coefficient value
print("Model intercept value: ", linreg.intercept_) # model intercept value
```
