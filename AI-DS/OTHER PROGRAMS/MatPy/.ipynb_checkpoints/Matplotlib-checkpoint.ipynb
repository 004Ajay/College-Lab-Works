# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.array([0, 6])
y = np.array([0, 25])

plt.plot(x, y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

```python
y_pts = np.array([3, 8, 1, 10])
x_pts = np.array([3, 8, 3, 8, 8, 3])

plt.plot(y_pts, marker='o', color='red')
plt.show()

plt.plot(x_pts, marker='x', color='green', linestyle = 'dotted')
plt.show()
```

```python
ypt = np.array([35, 25, 25, 15])
mylabels = ['1st year', '2nd year', '3rd year', '4th year']
myexplode = [0.1, 0, 0, 0]

plt.pie(ypt, labels=mylabels, explode=myexplode)
plt.show()
```
