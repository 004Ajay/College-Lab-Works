```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('car_details.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
sns.pairplot(df, x_vars = ['selling_price', 'km_driven'], y_vars = 'year',
             height=10, aspect=0.7, kind='reg')
```

```python
# Doing Linear Regression

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

X = df[['km_driven']] # can be replaced selling_price or km_driven
y = df.year

# splitting dataset into Train & Test Set
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_train.shape)

linreg = LinearRegression()
linreg.fit(X_train, y_train)

# Interpretting Model Coefficients
print("Model coefficient value: ", linreg.coef_) # model coefficient value
print("Model intercept value: ", linreg.intercept_) # model intercept value
```
