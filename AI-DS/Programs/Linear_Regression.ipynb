```python
from sklearn.datasets import load_iris

iris = load_iris()

X, y = iris.data, iris.target # Features & Target

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.33, random_state= 101)

# - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - - -

from sklearn.linear_model import LinearRegression

lr = LinearRegression().fit(X_train, y_train)
```

```python
def conv(species):
    if species == 0: return 'SETOSA'
    elif species == 1: return 'VERSICOLOR'
    elif species == 2: return 'VIRGINICA'
```

```python
# [4.3, 3. , 1.1, 0.1] - Setosa flower
# [6.3, 2.3, 4.4, 1.3] - Versicolor flower 
# [6.8, 3.2, 5.9, 2.3] - Virginica flower
pred = int(lr.predict([[4.4, 2.9, 1.4, 0.2]])) # sample values of 4 features is given as 1D Array
species = conv(pred)
print(f'Predicted value: {pred}\nPrediction: {species} flower of iris species')
```

## Visualizing Iris Dataset before & after Linear Regression

```python
import pandas as pd

iris_df = pd.DataFrame(data= iris.data, columns= iris.feature_names)
iris_df
```

```python
target_df = pd.DataFrame(data= iris.target, columns= ['species'])
target_df['species'] = target_df['species'].apply(conv)
target_df
```

```python
iris_df = pd.concat([iris_df, target_df], axis= 1)
iris_df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# scatter plot of sepal length & sepal width
sns.scatterplot(x='sepal length (cm)', y='sepal width (cm)', hue='species', data=iris_df)
plt.show()
```

```python
# scatter plot of sepal length & sepal width
sns.regplot(x='sepal length (cm)', y='sepal width (cm)', data=iris_df)
plt.show()
```
